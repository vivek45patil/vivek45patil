```python
X=df.drop("Outcome",axis=1)
Y=df.Outcome
```

```python
print(X)
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
sc=StandardScaler()
scaled=sc.fit_transform(X)
```

```python
print(scaled)
```

```python
X=scaled

print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print("---------------------------------------------------------------------")
print(X_test)
```

```python
svm_class=svm.SVC(kernel='linear')

svm_class.fit(X_train,Y_train)
```

```python
X_train_pred=svm_class.predict(X_train)
X_train_accuracy=accuracy_score(X_train_pred,Y_train)
print("Accuracy = ",X_train_accuracy)
```

```python
X_test_pred=svm_class.predict(X_test)
X_test_accuracy=accuracy_score(X_test_pred,Y_test)
print("Accuracy score for X test = ",X_test_accuracy)
```

```python
values=(2,120,72,19,175,28.5,0.287,56)
data1=np.asarray(values)

#reshape array to prdict for one instance

data=data1.reshape(1,-1)
scaled_data=sc.transform(data)
print(scaled_data)

prediction=svm_class.predict(scaled_data)
print(prediction)

if prediction[0]==1:
  print(" The Person is Diabetic")
else:
  print(" The Person is not Diabetic ")
```
